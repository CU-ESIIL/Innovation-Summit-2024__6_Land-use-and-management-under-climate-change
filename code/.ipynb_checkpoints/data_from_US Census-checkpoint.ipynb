```python
# Specify the directory containing the CSV files
directory_path = '/home/jovyan/data-store/home/nilangakoon/LCMAP/vegetation'

# Initialize an empty list to store DataFrames
dataframes = []

# Iterate over each file in the directory
for filename in os.listdir(directory_path):
    # Check if the file is a CSV file
    if filename.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Append the DataFrame to the list of DataFrames
        dataframes.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('/home/jovyan/data-store/home/nilangakoon/LCMAP/developed_new/combined_vegetation.csv', index=False)
```

```python
# Load the shapefile
gdf = gpd.read_file('/home/jovyan/data-store/home/nilangakoon/LCMAP/CA_counties.shp')

# Load the CSV file
df = pd.read_csv('/home/jovyan/data-store/home/nilangakoon/LCMAP/developed_new/combined_vegetation.csv')
```

```python
# Convert the common ID columns to string in both the DataFrame and GeoDataFrame
gdf['GEOID'] = gdf['GEOID'].astype(str)
df['GEOID'] = df['GEOID'].astype(str)

# Now perform the join
joined_gdf = gdf.merge(df, on='GEOID', how='left')
```

```python
# Save the resulting GeoDataFrame to a new shapefile
joined_gdf.to_file('path_to_output_shapefile.shp', driver='ESRI Shapefile')
```
